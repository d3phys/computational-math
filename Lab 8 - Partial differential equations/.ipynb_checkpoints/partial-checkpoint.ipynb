```python
#
# Постановка задачи храним "внутри"
#
class Grid:
    def __init__(self, a, h, tau, x_min = 0.0, x_max = 1.0, t_min = 0.0, t_max = 1.0):
        self.a     = a
        self.h     = h
        self.tau   = tau
        self.x_min = x_min
        self.x_max = x_max
        self.t_min = t_min
        self.t_max = t_max

        self.x_nodes = int((self.x_max - self.x_min) / h)
        self.t_nodes = int((self.t_max - self.t_min) / tau)

        # Задаем сетку с начальными условиями
        self.grid       = np.zeros(shape = (self.t_nodes, self.x_nodes), dtype = np.float64)
        self.grid[0]    = self.initPhi_x(np.linspace(self.x_min, self.x_max, self.x_nodes))
        self.grid[:, 0] = self.initPhi_t(np.linspace(self.t_min, self.t_max, self.t_nodes))
    
    def array(self):
        return self.grid
    
    def initPhiFor_x(self, x): 
        __from = self.x_min + 0.25 * (self.x_max - self.x_min)
        __to   = self.x_min + 0.75 * (self.x_max - self.x_min)
        
        if ( x >= __from and x <= __to ):
            return 1.0
        else:
            return 0.0
        
    def initPhiFor_t(self, t):
        __from = self.t_min + 0.25 * (self.t_max - self.t_min)
        __to   = self.t_min + 0.75 * (self.t_max - self.t_min)

        if ( t >= __from and t <= __to ):
            __mid = (__to + __from) / 2.0            
            if ( t < __mid ):
                return (t - __from) / (__mid - __from)
            else:
                return (t - __to)   / (__mid - __to)
        else:
            return 0.0
            
    def initPhi_x(self, x_arr): 
        return np.array([self.initPhiFor_x(__x) for __x in x_arr])

    def initPhi_t(self, t_arr): 
        return np.array([self.initPhiFor_t(__t) for __t in t_arr])

#
# Вывести красивую сетку
#
def ShowGrid(grid):
    fig, ax = plt.subplots()
    im = ax.imshow(grid.array())
    ax.set_title(f'$h = {grid.h}, \\tau = {grid.tau}, a = {grid.a}$')
    
    fig.colorbar(im, ax = ax, label = '')
    
    plt.xlabel('$x$')
    plt.ylabel('$t$')
    
    plt.show()   
```

### __Disclaimer:__
__В условии начальный профиль я решил сделать задающим условия $\varphi(0, t)$ (cм. рис. ниже). Так получается прикольней :)__

```python
grid = Grid(1, 1, 1)

plt.figure(figsize=[8, 3], dpi=100)

plt.subplot(1, 2, 1)
x = np.linspace(grid.x_min, grid.x_max, 100)
plt.plot( x, grid.initPhi_x(x))
plt.ylabel('$\\varphi(x, 0)$')
plt.xlabel('$x$')

plt.subplot(1, 2, 2)
t = np.linspace(grid.t_min, grid.t_max, 100)
plt.plot( t, grid.initPhi_t(t))
plt.ylabel('$\\varphi(0, t)$')
plt.xlabel('$t$')

plt.tight_layout()
```

## Схема Куранта–Изаксона–Риса

<img src="КИР.png" width=700/>

```python
# Вычисляем узлы по явной схеме КИР
def SolveWithKIR(grid):
    grid_arr = grid.array()
    for n in range(0, grid_arr.shape[0] - 1):
        for m in range(1, grid_arr.shape[1]):
            grid_arr[n + 1][m] = grid_arr[n][m] - grid.tau * grid.a / grid.h * (grid_arr[n][m] - grid_arr[n][m - 1])
    ShowGrid(grid)
```

```python
SolveWithKIR(Grid(h = 0.5e-1, tau = 0.5e-1, a = 0.5))
```

```python
SolveWithKIR(Grid(h = 0.5e-3, tau = 0.5e-3, a = 0.5))
```

```python
SolveWithKIR(Grid(h = 0.5e-3, tau = 0.5e-3, a = 0.05))
```

## Схема Мак-Кормака

<img src="Дивергентная форма.png" width=700/>

<img src="Мак-Кормак.png" width=700/>

```python
def SolveWithMacCormack(grid):
    u = grid.array()
    for n in range(0, u.shape[0] - 1):
        for m in range(1, u.shape[1]):

            # Крайнюю точку обработаем уголком
            if m == u.shape[1] - 1:
                u[n + 1][m] = u[n][m] - grid.tau * grid.a / grid.h * (u[n][m] - u[n][m - 1])
                continue    
            
            # Предиктор
            u_m   = u[n][m]     - grid.tau / grid.h * grid.a * (u[n][m + 1] - u[n][m])
            u_m_1 = u[n][m - 1] - grid.tau / grid.h * grid.a * (u[n][m]     - u[n][m - 1])

            # Корректор
            u[n + 1][m] = 0.5 * (u[n][m] + u_m) - grid.tau / (2 * grid.h) * grid.a * (u_m - u_m_1)
            
    ShowGrid(grid)
```

```python
SolveWithMacCormack(Grid(h = 0.5e-1, tau = 0.5e-1, a = 0.5))
```

```python
SolveWithMacCormack(Grid(h = 0.5e-3, tau = 0.5e-3, a = 0.5))
```

```python
SolveWithMacCormack(Grid(h = 0.5e-3, tau = 0.5e-3, a = 0.05))
```

## Гибридная схема Федоренко

<img src="Федоренко.png" width=700/>

```python
def SolveWithFedorenko(grid):
    u = grid.array()
    for n in range(0, u.shape[0] - 1):
        for m in range(1, u.shape[1]):
    
            # Крайнюю точку обработаем уголком
            if m == u.shape[1] - 1:
                u[n + 1][m] = u[n][m] - grid.tau * grid.a / grid.h * (u[n][m] - u[n][m - 1])
                continue    
            
            # Схема Федоренко
            u[n + 1][m] = u[n][m] - grid.a * grid.tau / grid.h * (u[n][m] - u[n][m - 1]) \
                - 1/2 * (grid.a * grid.tau / grid.h - (grid.a * grid.tau / grid.h)**2) * (u[n][m - 1] - 2 * u[n][m] + u[n][m + 1]) 
            
    ShowGrid(grid)
```

```python
SolveWithFedorenko(Grid(h = 0.5e-1, tau = 0.5e-1, a = 0.5))
```

```python
SolveWithFedorenko(Grid(h = 0.5e-3, tau = 0.5e-3, a = 0.5))
```

```python
SolveWithFedorenko(Grid(h = 0.5e-3, tau = 0.5e-3, a = 0.05))
```

## TVD-схема

<img src="TVD1.png" width=700/>
<img src="TVD2.png" width=700/>
<img src="TVD3.png" width=700/>

```python
def SolveWithTVD(grid, limiter = lambda r_m: 1):
    u = grid.array()
    sigma = grid.a * grid.tau / grid.h
    
    for n in range(0, u.shape[0] - 1):
        for m in range(1, u.shape[1]):
    
            # Крайнюю точку обработаем уголком
            if m == u.shape[1] - 1:
                u[n + 1][m] = u[n][m] - grid.tau * grid.a / grid.h * (u[n][m] - u[n][m - 1])
                continue
    
            # Делаем схему монотонной
            epsilon = 1e-9 # Некоторое "малое число"
            r_m = (u[n][m] - u[n][m - 1] + epsilon) / (u[n][m + 1] - u[n][m] + epsilon)
    
            # TVD-схема
            f_plus  = 0.5 * limiter(r_m) * sigma * (1 - sigma) * (u[n][m + 1] - u[n][m])
            f_minus = 0.5 * limiter(r_m) * sigma * (1 - sigma) * (u[n][m] - u[n][m - 1])
            
            u[n + 1][m] = u[n][m] - sigma * (u[n][m] - u[n][m - 1]) - (f_plus - f_minus) 
            
    ShowGrid(grid)
```

```python
SolveWithTVD(Grid(h = 0.5e-1, tau = 0.5e-1, a = 0.5))
```

```python
SolveWithTVD(Grid(h = 0.5e-3, tau = 0.5e-3, a = 0.5))
```

```python
SolveWithTVD(Grid(h = 0.5e-3, tau = 0.5e-3, a = 0.05))
```

### Пробуем другой ограничитель

<img src="TVD-limiter.png" width=700/>

```python
def Phi(r_m):
    if r_m > 1:
        return min(2, r_m)
    elif 0 < r_m and r_m <= 1:
        return min(2 * r_m, 1)
    else:
        return 0
```

```python
SolveWithTVD(Grid(h = 0.5e-3, tau = 0.5e-3, a = 0.5), limiter = Phi)
```

```python
SolveWithTVD(Grid(h = 0.5e-3, tau = 0.5e-3, a = 0.05), limiter = Phi)
```
